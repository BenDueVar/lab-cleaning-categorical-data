```python
# I will remove the outliers that approximately are the 5% of the total data entries (9134) this means that 
# the upper limit will be adjusted accordingly after checking the 1.5 IQR.
```

```python
iqr = np.nanpercentile(continuous_df['total_claim_amount'],75) - np.nanpercentile(continuous_df['total_claim_amount'],25)
upper_limit = np.nanpercentile(continuous_df['total_claim_amount'],75) + 1.5*iqr
print(iqr)
print(upper_limit)

#checking how many rows are we cutting
extraordinary_points = continuous_df[continuous_df['total_claim_amount'] > 960]
print(len(extraordinary_points))
continuous_df = continuous_df[continuous_df['total_claim_amount']<=960]
len(continuous_df)
```

```python
customer_df = customer_df[customer_df['total_claim_amount']<=960]
#doing the same for the data base.
```

```python
# after total_claim_amount cut 100 values above the upper_limit were cut reducing the upper limit to 360 rows. I fill more comfortable without cutting these rows
iqr = np.nanpercentile(continuous_df['customer_lifetime_value'],75) - np.nanpercentile(continuous_df['customer_lifetime_value'],25)
upper_limit = np.nanpercentile(continuous_df['customer_lifetime_value'],75) + 1.5*iqr
print(iqr)
print(upper_limit)

#checking how many rows are we cutting
extraordinary_points = continuous_df[continuous_df['customer_lifetime_value'] > 22500]
len(extraordinary_points)
```

```python
#same as before the rows above the upper limit were reduced in around 170. 
iqr = np.nanpercentile(continuous_df['monthly_premium_auto'],75) - np.nanpercentile(continuous_df['monthly_premium_auto'],25)
upper_limit = np.nanpercentile(continuous_df['monthly_premium_auto'],75) + 1.5*iqr
print(iqr)
print(upper_limit)

#checking how many rows are we cutting
extraordinary_points = continuous_df[continuous_df['monthly_premium_auto'] > 170.5]
len(extraordinary_points)
```

Does any column contain alpha and numeric data? Decide how to clean it and do it now.

```python
#No. But number of months_since_last_claim and months_since_policy_inception can be treated as categorical if we state 4 percentils for each group, for example.

months_since_last_claim_labels = ['group_0-6', 'group_6-14', 'group_14-23', 'group_23-35']
numerical_df['months_since_last_claim'] = pd.qcut(numerical_df['months_since_last_claim'],4,labels=months_since_last_claim_labels)
categorical_df['months_since_last_claim']=numerical_df['months_since_last_claim']

months_since_policy_inception_labels = ['group_0-24', 'group_24-48', 'group_48-71', 'group_71-99']
numerical_df['months_since_policy_inception'] = pd.qcut(numerical_df['months_since_policy_inception'],4,labels=months_since_policy_inception_labels)
categorical_df['months_since_policy_inception']=numerical_df['months_since_policy_inception']
```

Would you choose to do anything else to clean or wrangle the categorical data? Comment your decisions and do it now.

```python
# 1. Customer Id I will remove it
# 2. States Washington and Nevada I will regroup them. Same with employment status, maybe it can be employed, unemployed and other.
# Same with vehicle class luxury classes and sport cars can be held as a unique class
```

```python
categorical_df = categorical_df.drop(['customer'],axis=1)
```

```python
categorical_df['vehicle_class'] = categorical_df['vehicle_class'].replace(['Sports Car','Luxury SUV','Luxury Car'], 'Luxury')
categorical_df['state'] = categorical_df['state'].replace(['Nevada','Washington'], 'Other')
categorical_df['employmentstatus'] = categorical_df['employmentstatus'].replace(['Medical Leave', 'Disabled', 'Retired'], 'Other')
```

```python
# let's try to see if the assumption on keeping policy is right
categorical_df.groupby('renew_offer_type')['policy'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.xlabel('renew_offer_type')
plt.ylabel('count')
plt.show()
# I see that each offer is more or less in the same proportion of policy. Therefore no extra info is added by the policy column regarding the renew offer type column
```

```python
categorical_df.groupby('vehicle_class')['education'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.xlabel('vehicle_class')
plt.ylabel('count')
plt.show()
# We knew it from before and luxury cars have a higher proportion of customers with high school or below education level. 
# Maybe because they jump into the market before and this make them more likely to be young and "wealthy-enough" to buy a luxury car
```

```python
categorical_df.groupby('coverage')['employmentstatus'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.xlabel('coverage')
plt.ylabel('count')
plt.show()
# I was expecting that the premium coverage is significantly reduced (in proportion and absolute number) among unemployed customers
```

```python
categorical_df.groupby('renew_offer_type')['sales_channel'].value_counts().unstack().plot(kind='bar', stacked=True)
plt.xlabel('renew_offer_type')
plt.ylabel('count')
plt.show()
# Offer number two was given through the web more than compare to other offers. We may think that that web campaign was very succesful
```
